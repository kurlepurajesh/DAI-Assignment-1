```python

# Categorical values check
cat_cols = ['Item', 'Payment Method', 'Location']
print("\nUnique Categorical Values:")
for col in cat_cols:
    print(f"\n{col}:")
    display(df[col].value_counts(dropna=False))
```

Handle Missing Values


```python
# Replace placeholder strings with NaN
df.replace(['ERROR', 'UNKNOWN', '', ' '], np.nan, inplace=True)
```

```python
# Convert numerical columns to proper type
numeric_cols = ['Quantity', 'Price Per Unit', 'Total Spent']
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
# Impute numerical columns
num_imputer = SimpleImputer(strategy='median')
df[numeric_cols] = num_imputer.fit_transform(df[numeric_cols])
```

```python
# Impute categorical columns
cat_imputer = SimpleImputer(strategy='most_frequent')
df[cat_cols] = cat_imputer.fit_transform(df[cat_cols])
```

```python
# Convert to datetime with more format flexibility
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'], errors='coerce', format='%Y-%m-%d')

# Remove rows with invalid dates
df = df[df['Transaction Date'].notna()]

# Extract useful date components
df['Year'] = df['Transaction Date'].dt.year
df['Month'] = df['Transaction Date'].dt.month
df['Day'] = df['Transaction Date'].dt.day
df['Weekday'] = df['Transaction Date'].dt.weekday
df['Hour'] = df['Transaction Date'].dt.hour
```

Handle Duplicates

```python
print("Duplicates Before Removal:", df.duplicated().sum())
df.drop_duplicates(inplace=True)
print("Duplicates After Removal:", df.duplicated().sum())
```

Outlier Treatment

```python
def handle_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Cap outliers
    data[column] = np.where(data[column] < lower_bound, lower_bound,
                          np.where(data[column] > upper_bound, upper_bound, data[column]))
    return data
```

```python
# Visualize before treatment
plt.figure(figsize=(15,5))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(1,3,i)
    sns.boxplot(data=df, y=col)
    plt.title(f'Before Treatment: {col}')
plt.tight_layout()
plt.show()
```

```python
# Apply outlier treatment
for col in numeric_cols:
    df = handle_outliers(df, col)
```

```python
# Visualize after treatment
plt.figure(figsize=(15,5))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(1,3,i)
    sns.boxplot(data=df, y=col)
    plt.title(f'After Treatment: {col}')
plt.tight_layout()
plt.show()
```

```python
# Clean Item names
df['Item'] = df['Item'].str.strip().str.title()
df['Item'] = df['Item'].replace({
    'Cookies': 'Cookie',
    'Juices': 'Juice',
    'Cofee': 'Coffee'
})
```

```python
# Standardize Payment Method
df['Payment Method'] = df['Payment Method'].str.strip().str.title()
df['Payment Method'] = df['Payment Method'].replace({
    'Creditcard': 'Credit Card',
    'Digitalwallet': 'Digital Wallet'
})
```

```python
# Clean Location
df['Location'] = df['Location'].str.strip().str.title()
df['Location'] = df['Location'].replace({
    'Takeaway': 'Takeout',
    'Instore': 'In-Store'
})
```

Final Validation

```python
print("\nFinal Data Structure:")
display(df.info())
```

```python
print("\nMissing Values After Cleaning:")
display(df.isnull().sum())
```

```python
print("\nUnique Categorical Values After Cleaning:")
for col in cat_cols:
    print(f"\n{col}:")
    display(df[col].value_counts())
```

```python
print("\nSummary Statistics After Cleaning:")
display(df.describe())
```

```python
df.to_csv('cleaned_cafe_sales.csv', index=False)
print("Cleaned dataset saved successfully!")
```
